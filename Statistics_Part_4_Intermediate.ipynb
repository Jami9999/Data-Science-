```python
import numpy as N
Marks = N.random.randint (1 , 100 , 40)
```

```python
import pandas as P
DFM = P.DataFrame (Marks , columns = ["Student Marks"])
print(DFM)
```

```python
def Result (X) :
  if X >= 80 :
    return "A+"
  elif X >= 75 and X <= 79 :
    return "A"
  elif X >= 70 and X <= 74 :
    return "A-"
  elif X >= 65 and X <= 69 :
    return "B+"
  elif X >= 60 and X <= 64 :
    return "B"
  elif X >= 55 and X <= 59 :
    return "C+"
  elif X >= 50 and X <= 54 :
    return "C"
  elif X >= 45 and X <= 49 :
    return "D+"
  elif X >= 40 and X <= 44 :
    return "D"
  else :
    return "F"
DFM ["Student Result"] = DFM["Student Marks"].apply (Result)
print(DFM)
```

```python
Sample_1 = DFM.groupby("Student Result").sample ()
print(Sample_1)
print("\n")
print(Sample_1.head ())
```

```python
Sample_2 = DFM.groupby("Student Result" , group_keys = False).sample ()
print(Sample_2)
print("\n")
print(Sample_2.head ())
print("\n")
Sample_3 = DFM.groupby("Student Result" , group_keys = True).sample ()
print(Sample_3)
print("\n")
print(Sample_3.head ())
```

```python
Mean_1 = []
for i in range (40) :
  Sample_4 = N.random.choice (Marks , 40)
  Mean_1.append (N.mean (Sample_4))
print(Mean_1)
```

```python
import matplotlib.pyplot as PLOT
PLOT.hist (Mean_1)
PLOT.show ()
```

```python
import seaborn as SEA
SEA.histplot (Mean_1)
PLOT.show ()
```

```python
Marks_Large_S = N.random.randint (1 , 100 , 80)
Mean_Marks_Large_S = N.mean (Marks_Large_S)
print(Mean_Marks_Large_S)
```

```python
Mean_2 = []
for i in range (80) :
  Sample_5 = N.random.choice (Marks_Large_S , 80)
  Mean_2.append (N.mean (Sample_5))
print(Mean_2)
print("\n")
PLOT.hist (Mean_2)
PLOT.show ()
print("\n")
SEA.histplot (Mean_2)
PLOT.show ()
```
