```python
H = D["Home Team"].value_counts()
A = D["Away Team"].value_counts()
Total = H + A
print(Total)
```

```python
TOTAL_SORT = Total.sort_values(ascending = False).head (5)
print(TOTAL_SORT)
```

```python
import matplotlib.pyplot as PLOT
```

```python
PLOT.bar (TOTAL_SORT.index , TOTAL_SORT.values)
PLOT.title ("Top 5 Teams")
PLOT.xlabel ("Teams")
PLOT.ylabel ("Matches")
PLOT.show ()
```
