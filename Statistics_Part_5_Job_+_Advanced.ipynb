```python
import warnings
warnings.filterwarnings('ignore')
import numpy as N
from scipy import stats
```

```python
OLD = [60 , 62 , 65 , 63]
NEW = [70 , 75 , 78 ,72]
TT_1 , PV_1 = stats.ttest_ind (OLD , NEW)
print(PV_1)
print(PV_1 / 2)
print(TT_1)
```

```python
OM = [5 , 6 , 7 , 5]
NM = [2 , 3 , 2 , 1]
TT_2 , PV_2 = stats.ttest_ind (OM , NM)
print(TT_2)
print("\n")
print(PV_2)
print("\n")
print(PV_2 / 2)
```

```python
Transactions = N.random.normal (100 , 5 , 200)
Mean = N.mean (Transactions)
STD = N.std (Transactions)
SQRT = N.sqrt (Transactions)
Z_Score = (Mean - 100) / (STD / SQRT)
print(Mean)
print("\n")
print(STD)
print("\n")
print(SQRT)
print("\n")
print(Z_Score)
```
