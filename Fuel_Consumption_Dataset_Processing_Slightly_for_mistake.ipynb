```python
from sklearn.preprocessing import LabelEncoder
```

```python
D["Make"] = LabelEncoder ().fit_transform (D["Make"])
D["Model"] = LabelEncoder ().fit_transform (D["Model"])
D["Vehicle Class"] = LabelEncoder ().fit_transform (D["Vehicle Class"])
D["Transmission"] = LabelEncoder ().fit_transform (D["Transmission"])
D["Fuel Type"] = LabelEncoder ().fit_transform (D["Fuel Type"])
D["Engine Group"] = LabelEncoder ().fit_transform (D["Engine Group"])
```

```python
D = D.drop_duplicates ()
```

```python
D["Efficiency Ratio City"] = D["Fuel Efficiency City"] / D["Engine Size(L)"]
```

```python
D["Efficiency Ratio Highway"] = D["Fuel Efficiency Highway"] / D["Engine Size(L)"]
```

```python
D["Efficiency Ratio Combined"] = D["Fuel Efficiency Combined"] / D["Engine Size(L)"]
```

```python
print(D.columns)
```

```python
def Emission_Rate_CO2_Level (X) :
  if X >= 6 :
    return 1
  elif X == 5 :
    return 0.83
  elif X == 4 :
    return 0.66
  elif X == 3 :
    return 0.49
  elif X == 2 :
    return 0.32
  else :
    return 0
D["Emission Rate CO2 level"] = D["CO2 Rating"].apply (Emission_Rate_CO2_Level)
```
