```python
Data_1 = {
    "Product" : ["A" , "B" , "C"] ,
    "Price" : [100 , 200 , 150] ,
    "Quality" : [2 , 3 , 1]
}
DF_1 = P.DataFrame (Data_1)
print(DF_1)
print("\n")
DF_1 ["Total Sales"] = DF_1 ["Price"] * DF_1 ["Quality"]
print(DF_1["Total Sales"])
print("\n")
print(DF_1)
print("\n")
print(DF_1["Total Sales"].mean ())
print("\n")
```
