```python
import pandas as P
```

```python
D = P.read_csv ("E-commerce Sales Dataset (Large Sample).csv")
```

```python
print(D.columns)
```

```python
D["Revenue"] = D["Price"] * D["Quantity"]
print(D["Revenue"])
```

```python
TP = D.groupby("Product")["Quantity"].sum ().sort_values(ascending = False)
print(TP)
```

```python
CP = D.groupby("City")["Revenue"].sum ().sort_values(ascending = False)
print(CP)
```

```python
J_F_C = D[D["Date"].isin([1 , 2])] / D.groupby("Date")["Revenue"].sum ()
print(J_F_C)
```

```python
TC = D.groupby("Customer_ID")["Quantity"].sum ().sort_values (ascending = False)
print(TC)
```
