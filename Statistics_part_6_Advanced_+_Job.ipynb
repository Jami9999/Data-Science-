```python
import pandas as P
```

```python
Data = NUM.random.randint (1 , 6 , 10)
DF_D = P.DataFrame (Data , columns = ["Dice Rolling"])
print(DF_D)
```

```python
Mean = DF_D.rolling(10).mean ()
print(Mean)
```

```python
STD = DF_D.rolling (20).std ()
print(STD)
```

```python
U = Mean + 1.96 * STD
L = Mean - 1.96 * STD
print(U)
print("\n")
print(L)
```

```python
print(U.head ())
print("\n")
print(L.head ())
print("\n")
print(U.tail ())
print("\n")
print(L.tail ())
```

```python
A = NUM.random.randint (1 , 40 , 20)
B = NUM.random.randint (1 , 40 , 80)
MA = NUM.mean (A)
MB = NUM.mean (B)
STDA = NUM.std (A)
STDB = NUM.std (B)
VARA = NUM.var (A)
VARB = NUM.var (B)
SE = NUM.sqrt ((VARA / len (A)) + (VARB / len (B)))
DIFF = MA - MB
U = DIFF + 1.96 * SE
L = DIFF - 1.96 * SE
print(U)
print("\n")
print(L)
print("\n")
print(MA)
print("\n")
print(MB)
print("\n")
print(STDA)
print("\n")
print(STDB)
print("\n")
```

```python
P = 15 / 500
FC = NUM.sqrt ((P * (1 - P)) / P)
Margin_Error = 1.96 * FC
U = P - Margin_Error
L = P + Margin_Error
print(U)
print(L)
print(Margin_Error)
```
