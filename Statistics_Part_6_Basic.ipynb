```python
import numpy as NUM
from scipy import stats
import warnings
warnings.filterwarnings("ignore")
```

```python
Mean = 60
STD = 10
N = 50
SQRT_N = NUM.sqrt (N)
Z = 1.96
Margin_Error = Z * (STD / SQRT_N)
CI_L = Mean - Margin_Error
CI_H = Mean + Margin_Error
print(CI_L)
print("\n")
print(CI_L)
```

```python
Z = stats.norm.ppf (0.95)
print(Z)
```

```python
Z = stats.norm.ppf (0.99)
print(Z)
```

```python
STD = 20
N = 64
Z = 1.96
SQRT = NUM.sqrt (N)
ME = Z * (STD / SQRT)
print(ME)
```

```python
Data = [10 , 12 , 14 , 15 , 13]
Mean = NUM.mean (Data)
STD = NUM.std (Data)
N = len (Data)
Z = 1.96
SQRT = NUM.sqrt (N)
ME = Z * (STD / SQRT)
CI_L = Mean - ME
CI_H = Mean + ME
print(CI_L)
print("\n")
print(CI_H)
```

```python
ST = stats.t.ppf (0.97 , df = 7)
print(ST)
```
