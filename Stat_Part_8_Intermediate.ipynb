```python
import pandas as P
import numpy as NUM
import matplotlib.pyplot as PLOT
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
```

```python
Data = {
    "Ads" : [100 , 200 , 300 , 400 , 500] ,
    "Price" : [10 , 9 , 8 , 7 , 6] ,
    "Sales" : [20 , 25 , 30 , 35 , 40]
}
DF = P.DataFrame (Data)
print(DF)
print("\n")
X = DF [["Ads" , "Price"]]
Y = DF["Sales"]
print(X)
print("\n")
print(Y)
M = LinearRegression ()
F = M.fit (X , Y)
print(M)
print("\n")
print(F)
PRE_1 = M.predict ([[600 , 3]])
print("\n")
print(PRE_1)
```

```python
from sklearn.metrics import r2_score
PRE_2 = M.predict (X)
R2 = r2_score (Y , PRE_2)
print(R2 * 100 , "%")
```

```python
from scipy.stats import f_oneway
C_1 = [100 , 110 , 120]
C_2 = [130 , 140 , 135]
C_3 = [90 , 95 , 100]
F , PV = f_oneway (C_1 , C_2 , C_3)
print(F)
print(PV)
if PV < 0.05 :
  print("Significant Difference")
else :
  print("No Significant Difference")
```

```python
P_spam = 0.4
P_N_spam = 0.6
P_W_spam = 0.8
P_N_W_spam = 0.2
PR = (P_spam * P_W_spam) / ((P_spam * P_W_spam) + (P_N_spam * P_N_W_spam))
print(PR * 100 , "%")
```

```python
from scipy.stats import binom
PN = 5
N = 20
Prob_1 = 0.3
Prob = binom.pmf (PN , N , Prob_1)
print(Prob * 100 , "%")
```
