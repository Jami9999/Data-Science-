```python
X = D[["League_Rank_1" , "League_Rank_2" , "Goals_Diff_1" , "Goals_Diff_2" , "Odds_Home"]]
Y = D["Results_1"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train , X_Test , Y_Train , Y_Test = train_test_split (X , Y , test_size = 0.2 , random_state = 42)
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

```python
NULL_X_Train = X_Train.isnull().sum()
print(NULL_X_Train)
NULL_X_Test = X_Test.isnull().sum()
print(NULL_X_Test)
NULL_Y_Train = Y_Train.isnull().sum()
print(NULL_Y_Train)
NULL_Y_Test = Y_Test.isnull().sum()
print(NULL_Y_Test)
```

```python
Fill_NULL_X_Train = X_Train.fillna(X_Train.mean ())
print(Fill_NULL_X_Train)
Fill_NULL_X_Test = X_Test.fillna(X_Test.mean ())
print(Fill_NULL_X_Test)
Fill_NULL_Y_Train = Y_Train.fillna(Y_Train.mean ())
print(Fill_NULL_Y_Train)
Fill_NULL_Y_Test = Y_Test.fillna(Y_Test.mean ())
print(Fill_NULL_Y_Test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
M_1 = DecisionTreeClassifier()
F_1 = M_1.fit(X_Train , Y_Train)
```

```python
PRE = F_1.predict (X_Test)
print(PRE)
```

```python
from sklearn.metrics import r2_score , mean_squared_error , mean_absolute_error
import numpy as NUM
```

```python
R2 = r2_score (Y_Test , PRE)
print(R2 * 100 , "%")
MSE = mean_squared_error (Y_Test , PRE)
print(MSE)
MAE = mean_absolute_error (Y_Test , PRE)
print(MAE)
RMSE = NUM.sqrt (MSE)
print(RMSE)
```
