```python
import pandas as P
import numpy as N
Student_ID = N.arange (1 , 1001)
DFS = P.DataFrame (Student_ID)
SAMPLE = DFS.sample (n = 363 , random_state = 42)
print(DFS)
print("\n")
print(SAMPLE.head ())
print("\n")
print(SAMPLE)
```

```python
import numpy as N
Marks = N.random.randint (63 , 101 , 363)
Sample_Mean = N.random.choice (Marks , 263)
print(Sample_Mean.mean ())
```

```python
import numpy as N
Marks = N.random.randint (40 , 100 , 1000)
Sample = N.random.choice (Marks , 100)
print(Marks.mean ())
print(Sample.mean ())
```

```python
Multiple_Sample = N.random.choice (Marks , size = (50 , 100))
SM = N.mean (Multiple_Sample , axis = 1)
print(SM)
print("\n")
```

```python
Small_Sample = N.random.choice (Marks , 10)
Big_Sample = N.random.choice (Marks , 100)
SM = N.mean (Small_Sample)
BM = N.mean (Big_Sample)
print(SM)
print("\n")
print(BM)
```

```python
DFM = P.DataFrame (Marks)
print(DFM)
print("\n")
RS = DFM.sample ()
print(RS)
print("\n")
RS_Row = DFM.sample (frac = 0.1)
print(RS_Row)
print("\n")
print(RS_Row.head ())
```

```python
Sample_Replacing = N.random.choice (Marks , 100 , replace = True)
Sample_Not_R = N.random.choice (Marks , 100 , replace = False)
print(Sample_Replacing)
print("\n")
print(Sample_Replacing [:10])
print("\n")
print(Sample_Not_R)
print("\n")
print(Sample_Not_R [:10])
```
