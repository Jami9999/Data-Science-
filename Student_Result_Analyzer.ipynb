```python
import pandas as P
```

```python
D = {
    "Name" : ["Arafat" , "Nusrat" , "Siam" , "Tanjim" , "Mim" , "Rakib"] ,
    "Marks" : [72 , 88 , 65 , 91 , 79 , 84]
}
```

```python
DF = P.DataFrame(D)
```

```python
Average_Marks = DF["Marks"].mean ()
print("Average Marks of students: " , Average_Marks)
```

```python
Highest_Marks = DF["Marks"].max ()
print("Highest Marks of students: " , Highest_Marks)
```

```python
Lowest_Marks = DF["Marks"].min ()
print("Lowest Marks of students: " , Lowest_Marks)
```

```python
if (DF["Marks"] >= 80).any () :
  print("Above Number 80 students get A+")
```
