```python
import matplotlib.pyplot as PLOT
import numpy as NUM
import pandas as P
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
Data_1 = {
    "Month" : ["Jan" , "Feb" , "Mar"] ,
    "Sales" : [1000 , 1500 , 1200]
}
DF_1 = P.DataFrame (Data_1)
print(DF_1)
print("\n")
PLOT.plot (DF_1["Month"] , DF_1["Sales"])
PLOT.show ()
print ("\n")
PLOT.plot (DF_1["Month"] , DF_1["Sales"])
PLOT.show ()
```

```python
Data_2 = {
    "Category" : ["Electronics" , "Clothing" , "Electronics"] ,
    "Sales" : [1000 , 500 , 1500]
}
DF_2 = P.DataFrame (Data_2)
print(DF_2)
print("\n")
G = DF_2.groupby ("Category")["Sales"].sum ()
print(G)
```

```python
Data_3 = {
    "Group" : ["A" , "B" , "A" , "B"] ,
    "Conversion" : [1 , 0 , 1 , 1]
}
DF_3 = P.DataFrame (Data_3)
print(DF_3)
print("\n")
G = DF_3.groupby ("Group")["Conversion"].sum ()
print(G)
```

```python
Data_4 = {
    "Speed" : [40 , 60 , 80 , 100] ,
    "Fuel" : [8 , 6 , 5 , 4]
}
DF_4 = P.DataFrame (Data_4)
print(DF_4)
C = DF_4.corr ()
print("\n")
print(C)
```
