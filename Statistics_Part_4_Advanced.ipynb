```python
import numpy as N
Marks = N.random.randint (1 , 100 , 80)
Mean_1 = N.mean (Marks)
print(Mean_1)
```

```python
E = []
for i in range (80) :
  Sample_1 = N.random.choice (Marks , 80)
  Error = (N.mean(Sample_1) - Mean_1)
  E.append(Error)
print(E)
print("\n")
print(N.mean(E))
```

```python
BM = []
for i in range (80) :
  Sample_2 = N.random.choice (Marks , 80 , replace =True)
  Mean_Sample_2 = N.mean(Sample_2)
  BM.append(Mean_Sample_2)
print(BM)
```

```python
import pandas as P
DFM = P.DataFrame(Marks , columns = ["Mark"])
print(DFM)
```

```python
def Result (X) :
  if X >= 80 :
    return "A+"
  elif X >= 75 and X <= 79 :
    return "A"
  elif X >= 70 and X <= 74 :
    return "A-"
  elif X >= 65 and X <= 69 :
    return "B+"
  elif X >= 60 and X <= 64 :
    return "B"
  elif X >= 55 and X <= 59 :
    return "C+"
  elif X >= 50 and X <= 54 :
    return "C"
  elif X >= 45 and X <= 49 :
    return "D+"
  elif X >= 40 and X <= 44 :
    return "D"
  else :
    return "F"
DFM["R"] = DFM["Mark"].apply (Result)
```

```python
print(DFM)
```

```python
RS = DFM.sample ()
SS = DFM.groupby ("R").sample ()
print(RS)
print("\n")
print(SS)
```

```python
BS = DFM [DFM ["Mark"] > 90]
print(BS["Mark"].mean ())
print("\n")
print(DFM["Mark"].mean ())
```
