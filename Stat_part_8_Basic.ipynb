```python
import pandas as P
import matplotlib.pyplot as PLOT
import warnings
warnings.filterwarnings("ignore")
import numpy as NUM
from sklearn.linear_model import LinearRegression
```

```python
Data = {
    "Experience" : [1 , 2 , 3 , 4 , 5] ,
    "Education" : [12 , 14 , 16 , 18 , 20] ,
    "Salary" : [20000 , 25000 , 30000 , 35000 , 40000]
}
DF = P.DataFrame (Data)
print(DF)
print("\n")
X = DF [["Experience" , "Education"]]
Y = DF ["Salary"]
print(X)
print("\n")
print(Y)
M = LinearRegression ()
F = M.fit (X , Y)
print(M)
print("\n")
print(F)
print("\n")
PRE = M.predict ([[6 , 22]])
print(PRE[0])
print("\n")
```

```python
COE = M.coef_
INT = M.intercept_
print(COE)
print("\n")
print(INT)
```

```python
from scipy.stats import f_oneway
G_1 = [20 , 22 , 23]
G_2 = [30 , 32 , 31]
G_3 = [40 , 42 , 41]
F_S , P_V = f_oneway (G_1 , G_2 , G_3)
print(F_S)
print(P_V)
if P_V < 0.05 :
  print("Significant Difference")
else :
  print("No Significant Difference")
```

```python
from scipy.stats import binom
PN = 6
N = 10
PRO = 0.5
BD_P = binom.pmf (PN , N , PRO)
print(BD_P * 100 , "%")
```

```python
from scipy.stats import poisson
PN = 5
LAM = 2
PD_P = poisson.pmf (PN , LAM)
print(PD_P * 100 , "%")
```
