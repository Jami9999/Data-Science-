```python
import pandas as P
```

```python
D = {
    "Name" : ["A" , "B" , "C" , "D" , "E"] ,
    "Marks" : [85 , 90 , None , 70 , 60] ,
    "Department" : ["CSE" , "EEE" , "CSE" , "BBA" , "EEE"] ,
    "Gender" : ["M" , "F" , "M" , "F" , "M"]
}
```

```python
DF = P.DataFrame(D)
print(DF)
```

```python
SF = DF.to_csv("Marks.csv" , index = False)
```

```python
Filling_Value = DF["Marks"].fillna(DF["Marks"].mean () , inplace = True)
print(Filling_Value)
```

```python
print(DF)
```

```python
Drop_Duplicate = DF.drop_duplicates ()
print(Drop_Duplicate)
```

```python
Finding_Greater_80 = DF[DF["Marks"] > 80]
print(Finding_Greater_80)
```

```python
print(DF["Marks"])
```

```python
Grouping_Marks = DF.groupby("Department")["Marks"].mean ()
print(Grouping_Marks)
```

```python
Counting_Gender = DF.groupby("Gender")["Name"].count ()
print(Counting_Gender)
```

```python
def R (X) :
  if X >= 80 :
    return "A+"
  elif X >= 75 and X <= 79 :
    return "A"
  elif X >= 70 and X <= 74 :
    return "A-"
  elif X >= 65 and X <= 69 :
    return "B+"
  elif X >= 60 and X <= 65 :
    return "B"
  elif X >= 55 and X <= 59 :
    return "C+"
  elif X >= 50 and X <= 54 :
    return "C"
  elif X >= 45 and X <= 49 :
    return "B+"
  elif X >= 40 and X <= 45 :
    return "B"
  else :
    return "F"
DF["Result"] = DF["Marks"].apply (R)
print(DF)
```

```python
Sorting_Marks = DF.sort_values(by = "Marks" , ascending = False)
print(Sorting_Marks)
```

```python
Top_3_Students = DF.sort_values(by = "Marks" , ascending = False).head (3)
print(Top_3_Students)
```
