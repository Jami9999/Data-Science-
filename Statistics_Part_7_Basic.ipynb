```python
Data = {
    "Junk Food" : [1 , 2 , 3 , 4 , 5] ,
    "Health Score" : [90 , 80 , 70 , 60 , 50]
}
DF = P.DataFrame (Data)
print(DF)
print("\n")
print(DF.corr ())
```

```python
PLOT.scatter (DF["Junk Food"] , DF["Health Score"])
PLOT.title ("Junk Food vs Health Score")
PLOT.xlabel ("Junk Food")
PLOT.ylabel ("Health Score")
PLOT.show ()
```

```python
Data = {
    "X" : [1 , 2 , 3 , 4 , 5] ,
    "Y" : [20000 , 30000 , 40000 , 50000 , 60000]
}
DF = P.DataFrame (Data)
print(DF)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_T = DF[["X"]]
Y_T = DF["Y"]
X_Train , X_Test , Y_Train , Y_Test = train_test_split (X_T , Y_T , test_size = 0.2 , random_state = 42)
```

```python
M = LinearRegression ()
M.fit (X_Train , Y_Train)
```

```python
PREDICT = M.predict (X_Test)
print(PREDICT)
```
