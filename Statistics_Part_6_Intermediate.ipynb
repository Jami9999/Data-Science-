```python
P = 0.63
N = 200
SQRT_N = NUM.sqrt (N)
SQRT_P = NUM.sqrt (P * (1 - P))
Z = 1.96
ME = Z * (SQRT_P / SQRT_N)
CIL = ME - Z
CIH = ME + Z
print(CIL)
print("\n")
print(CIH)
```
