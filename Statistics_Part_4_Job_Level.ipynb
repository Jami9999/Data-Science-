```python
import numpy as N
P_1 = N.random.randint (1 , 100 , 40)
S_1 = N.random.choice (P_1 , 30)
M_1 = N.mean (S_1)
print(M_1)
```

```python
STD = N.std (S_1)
ERROR = STD / N.sqrt (30)
CI_L = M_1 - 1.96 * ERROR
CI_H = M_1 + 1.96 * ERROR
print(ERROR)
print("\n")
print(CI_L)
print("\n")
print(CI_H)
```

```python
Marks = N.random.randint (1 , 100 , 40)
G_A = N.random.choice (Marks , 20)
G_B = N.random.choice (Marks , 20)
print(N.mean (G_A))
print("\n")
print(N.mean (G_B))
```
