```python
import pandas as P
import numpy as NUM
import matplotlib.pyplot as PLOT
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
Data_1 = {
    "Ads" : [10 , 20 , 30 , 40] ,
    "Sales" : [100 , 200 , 300 , 400]
}
DF_1 = P.DataFrame (Data_1)
print(DF_1)
print("\n")
X = DF_1[["Ads"]]
Y = DF_1["Sales"]
X_Train ,X_Test , Y_Train , Y_Test = train_test_split (X , Y , test_size = 0.07 , random_state = 42)
print(X_Train)
print("\n")
print(X_Test)
print("\n")
print(Y_Train)
print("\n")
print(Y_Test)
M_1 = LinearRegression ()
F_1 = M_1.fit (X_Train , Y_Train)
print(M_1)
print("\n")
print(F_1)
PRE_1 = M_1.predict (X_Test)
print(PRE_1)
print("\n")
R2 = r2_score (Y_Test , PRE_1)
print (R2)
print("\n")
PRE_1_S = M_1.predict ([[50]])
print(PRE_1_S)
```

```python
from sklearn.tree import DecisionTreeClassifier
Data_2 = {
    "Speed" : [40 , 80 , 60 , 100] ,
    "Rain" : [0 , 1 , 1 , 1] ,
    "Accident" : [0 , 1 , 0 , 1]
}
DF_2 = P.DataFrame (Data_2)
X = DF_2 [["Speed" , "Rain"]]
Y = DF_2 ["Accident"]
M_2 = DecisionTreeClassifier ()
F_2 = M_2.fit (X , Y)
PRE_2 = M_2.predict ([[90 , 1]])
print(PRE_2)
```

```python
from sklearn.cluster import KMeans
```

```python
Data_3 = {
    "Spending" : [100 , 200 , 300 , 400] ,
    "Visits" : [1 , 2 , 3 , 4]
}
DF_3 = P.DataFrame (Data_3)
X = DF_3 [["Spending"]]
Y = DF_3 ["Visits"]
M_3 = KMeans (n_clusters = 4)
F_3 = M_3.fit (X , Y)
PRE_3 = M_3.predict ([[583]])
print(DF_3)
print("\n")
print(PRE_3)
```

```python
from scipy import stats
A = [1 , 1 , 0 , 1 , 0]
B = [1 , 1 , 1 , 1 , 1]
TT_1 , PV_1 = stats.ttest_ind (A , B)
print(TT_1)
print(PV_1)
if PV_1 < 0.05 :
  print("Significant Difference")
else :
  print("No Significant Difference")
```

```python
Data_4 = {
    "Region" : ["North" , "South" , "North" , "South"] ,
    "Sales" : [1000 , 1500 , 2000 , 1200]
}
DF_4 = P.DataFrame (Data_4)
print(DF_4)
print("\n")
G_Sum = DF_4.groupby ("Region")["Sales"].sum ()
print(G_Sum)
BEST = G_Sum.idxmax ()
print("\n")
print(BEST)
```
