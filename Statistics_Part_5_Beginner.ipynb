```python
import numpy as N
```

```python
from scipy import stats
Data_1 = [45 , 50 , 55 , 60 , 48 , 52]
TT_1 , PV_1 = stats.ttest_1samp (Data_1 , 50)
print(TT_1)
print(PV_1)
if PV_1 < 0.05 :
  print("Null Hypothesos Rejected")
else :
  print("Null Hypothesis Accepted")
```

```python
from scipy import stats
Data_2 = [90 , 85 , 88 , 92]
TT_3 , PV_3 = stats.ttest_1samp (Data_2 , 100)
print(PV_3)
print(TT_3)
```

```python
Data_3 = N.random.normal (50 , 10 , 1000)
Mean_1 = N.mean (Data_3)
SQRT = N.sqrt(len (Data_3))
STD = N.std (Data_3)
Z_Score = (Mean_1 - 50) / (STD / SQRT)
print(Z_Score)
```
