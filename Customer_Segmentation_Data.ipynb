```python
import pandas as P
```

```python
D = P.read_csv ("Customer Segmentation Data.csv")
```

```python
print(D.columns)
```

```python
print(D.dtypes)
```

```python
NV = D.isnull ().sum ()
print(NV)
```

```python
DV = D.duplicated ().sum ()
print(DV)
```

```python
import matplotlib.pyplot as PLOT
PLOT.plot (D["Income"] , D["Spending"])
PLOT.xlabel ("Income")
PLOT.ylabel ("Spending")
PLOT.show ()
```

```python
from sklearn.cluster import KMeans
```

```python
KM = KMeans (n_clusters = 3)
D["Cluster"] = KM.fit_predict (D[["Income" , "Spending"]])
print(D["Cluster"])
```

```python
PLOT.scatter (D["Income"] , D["Spending"] , c = D["Cluster"])
PLOT.show()
```
