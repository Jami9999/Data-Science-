```python
import pandas as P
```

```python
D = {
    "Student" :["A" , "B" , "C" , "D" , "E" , "F" , "G"] ,
    "Marks" : [50 , 60 , 65 , 70 , 200 , 55 , 58]
}
DF = P.DataFrame(D)
print(DF)
```

```python
Mean = DF["Marks"].mean ()
print(Mean)
```

```python
Mode = DF["Marks"].mode ()
print(Mode)
```

```python
Median = DF["Marks"].median ()
print(Median)
```

```python
Minimum = DF["Marks"].min ()
Maximum = DF["Marks"].max ()
print(Minimum)
print(Maximum)
```

```python
Range = Maximum - Minimum
print(Range)
```

```python
Variance = DF["Marks"].var ()
STD = DF["Marks"].std ()
print(Variance)
print(STD)
```

```python
O = DF[(DF["Marks"] > Mean + 2 * STD) | (DF["Marks"] < Mean - 2 * STD)]
print(O)
```
