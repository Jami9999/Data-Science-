```python
import pandas as P
import matplotlib.pyplot as PLOT
import numpy as NUM
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error , r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
Data_1 = {
    "Experience" : [1 , 2 , 3 , 4 , 5 , 6 , 7 , 8] ,
    "Salary" : [15000 , 20000 , 25000 , 30000 , 35000 , 40000 , 45000 , 50000]
}
DF_1 = P.DataFrame (Data_1)
print(DF_1)
print("\n")
X = DF_1 [["Experience"]]
Y = DF_1 ["Salary"]
print(X)
print("\n")
print(Y)
X_Train , X_Test , Y_Train , Y_Test = train_test_split (X , Y , test_size = 0.2 , random_state = 42)
print(X_Train)
print("\n")
print(X_Test)
print("\n")
print(Y_Train)
print("\n")
print(Y_Test)
M_1 = LinearRegression ()
F_1 = M_1.fit (X_Train , Y_Train)
print(M_1)
print("\n")
print(F_1)
PRE_1 = F_1.predict (X_Test)
print("\n")
print(PRE_1)
print("\n")
PLOT.scatter (Y_Test , PRE_1)
PLOT.plot ([Y_Test.min () , Y_Test.max ()] , [Y_Test.min () , Y_Test.max ()])
PLOT.show ()
PRE_1_T = M_1.predict ([[10]])
print("\n")
print(PRE_1_T)
R2 = r2_score (Y_Test , PRE_1)
print("\n")
print(R2 * 100 , "%")


```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
Data_9 = {
    "D1" : [1 , 2 , 3 , 4 , 5 , 6] ,
    "D2" : [0 , 0 , 0 , 1 , 1 , 1]
}
DF_9 = P.DataFrame (Data_9)
print(DF_9)
print("\n")
X = DF_9 [["D1"]]
Y = DF_9 ["D2"]
print(X)
print("\n")
print(Y)
X_Train , X_Test , Y_Train , Y_Test = train_test_split (X , Y , test_size = 0.2 , random_state = 42)
print(X_Train)
print("\n")
print(X_Test)
print("\n")
print(Y_Train)
print("\n")
print(Y_Test)
print("\n")
MF = LogisticRegression ()
MS = RandomForestClassifier ()
MT = KNeighborsClassifier (n_neighbors = 2)
F_MF = MF.fit (X_Train , Y_Train)
F_MS = MS.fit (X_Train , Y_Train)
F_MT = MT.fit (X_Train , Y_Train)
print(MF)
print("\n")
print(MS)
print("\n")
print(MT)
print("\n")
print(F_MF)
print("\n")
print(F_MS)
print("\n")
print(F_MT)
print("\n")
PRE_MF = F_MF.predict (X_Test)
PRE_MS = F_MS.predict (X_Test)
PRE_MT = F_MT.predict (X_Test)
print(PRE_MF)
print("\n")
print(PRE_MS)
print("\n")
print(PRE_MT)
print("\n")
R2_MF = r2_score (Y_Test , PRE_MF)
R2_MS = r2_score (Y_Test , PRE_MS)
R2_MT = r2_score (Y_Test , PRE_MT)
print(R2_MF)
print("\n")
print(R2_MS)
print("\n")
print(R2_MT)
print("\n")
PLOT.hist (R2_MF , bins = 10)
PLOT.show ()
print("\n")
PLOT.hist (R2_MS , bins = 10)
PLOT.show ()
print ("\n")
PLOT.hist (R2_MT , bins = 10)
PLOT.show ()
print("\n")
MF_Score = MF.score (X_Test , Y_Test)
MS_Score = MS.score (X_Test , Y_Test)
MT_Score = MT.score (X_Test , Y_Test)
print (MF_Score)
print("\n")
print(MS_Score)
print("\n")
print(MT_Score)
print("\n")
def MODEL_C (MF_Score  , MS_Score , MT_Score) :
  if MF_Score > MS_Score :
    return "First Model is Best"
  elif MS_Score > MT_Score :
    return "Second Model is Best"
  elif MT_Score > MF_Score :
    return "Third Model is best"
  else :
    return "All Models Score is Same"
print(MODEL_C (MF_Score , MS_Score , MT_Score))
```
