```python
import numpy as NUM
import matplotlib.pyplot as PLOT
import pandas as P
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
Data = {
    "Experience" : [1 , 2 , 3 , 4 , 5] ,
    "Education" : [12 , 14 , 16 , 18 , 20] ,
    "Skills" : [2 , 3 , 4 , 5 , 6] ,
    "Salary" : [20000 , 26000 , 32000 , 38000 , 45000]
}
DF = P.DataFrame (Data)
print(DF)
print("\n")
X = DF [["Experience" , "Education" , "Skills"]]
Y = DF ["Salary"]
print(X)
print("\n")
print(Y)
X_Train , X_Test , Y_Train , Y_Test = train_test_split (X , Y , test_size = 0.2 , random_state = 42)
print(X_Train)
print("\n")
print(X_Test)
print("\n")
print(Y_Train)
print("\n")
print(Y_Test)
print("\n")
M = LinearRegression ()
F = M.fit (X_Train , Y_Train)
print(M)
print("\n")
print(F)
print("\n")
PRE_M = M.predict (X_Test)
print(PRE_M)
print("\n")
R2 = r2_score (Y_Test , PRE_M)
print(R2 * 100 , "%")
PRE_1 = M.predict ([[6 , 22 , 7]])
print("\n")
print(PRE_1)
```

```python
Residual = Y_Test - PRE_M
print(Residual)
```

```python
PLOT.scatter (PRE_M , Residual)
PLOT.axhline (0)
PLOT.title ("Residual Plot")
PLOT.show ()
```

```python
from scipy.stats import poisson
L = 0
PN = 3
Poisson = poisson.pmf (L , PN)
print(Poisson * 100 , "%")
```

```python
from scipy.stats import f_oneway
G_1 = [50 , 55 , 60]
G_2 = [65 , 70 , 75]
G_3 = [80 , 85 , 90]
F1 , PV = f_oneway (G_1 , G_2 , G_3)
print(PV)
print("\n")
print(F1)
print("\n")
if PV < 0.05 :
  print("Difference")
else :
  print("No Difference")
```
