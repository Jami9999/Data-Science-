```python
import pandas as P
import numpy as NUM


DF = P.read_csv("Nigeria Crash Data P.csv")


target_size = 1000
needed = target_size - len(DF)


DF_sampled = DF.sample(n = needed, replace = True, random_state = 42).reset_index(drop = True)


Numeric_cols = DF.select_dtypes(include=NUM.number).columns


noise = NUM.random.normal(
    loc=0,
    scale = DF[Numeric_cols].std().values * 0.02 ,
    size = DF_sampled[Numeric_cols].shape
)


DF_sampled[Numeric_cols] = DF_sampled[Numeric_cols] + noise


DF_final = P.concat([DF, DF_sampled], ignore_index = True)


DF_final = DF_final.sample(frac = 1, random_state = 42).reset_index(drop = True)


SF = DF_final.to_csv("Augmented Dataset of Nigeria Crash.csv", index = False)

print("Final shape:", DF_final.shape)
```
