```python
import pandas as P
```

```python
D = [10 , 20 , 20 , 30 , 30 , 30 , 40]
DF = P.DataFrame(D)
```

```python
Frequency = DF.value_counts()
print(Frequency)
```

```python
Frequency_Table = {
    "Index" : Frequency.index ,
    "Count_Of_Frequency" : Frequency.values
}
Frequency_Table_DF = P.DataFrame(Frequency_Table)
print(Frequency_Table_DF)
```

```python
Relative_Frequency = Frequency / len(DF)
print(Relative_Frequency)
```

```python
import matplotlib.pyplot as PLOT
```

```python
PLOT.hist (DF)
PLOT.title ("Frequency Distribution and Spread")
PLOT.xlabel ("Value")
PLOT.ylabel ("Frequency")
PLOT.show ()
```

```python
import numpy as N
```

```python
Count , Range = N.histogram (DF)
print(Range[N.argmax(Count)])
```

```python
C , R = N.histogram (DF)
print(R[N.argmax(C)])
```

```python
Mean = N.mean (DF)
Median = N.median (DF)
print(Mean , Median)
```

```python
def S (Mean , Median) :
  if Mean > Median :
    return "Positive Skew"
  else :
    return "Negative Skew or Symetric"
print(S(Mean , Median))
```

```python
Data_F_E_S = {
    "Employee ID" : [101 , 102 , 103 , 104 , 105] ,
    "Working Hours" : [10 , 8 , 12 , 7 , 9] ,
    "Sleep Hours" : [5 , 6 , 4 , 7 , 5] ,
    "Deadline Preassure" : ["High" , "Medium" , "High" , "Low" , "Medium"] ,
    "Stress Level" : ["High" , "Medium" , "High" , "Low" , "Medium"]
}
DF_F_E_S = P.DataFrame (Data_F_E_S)
print(DF_F_E_S)
```

```python
import matplotlib.pyplot as PLO
import seaborn as SE
print(DF_F_E_S)
print ("\n")
SE.histplot (DF_F_E_S["Working Hours"])
PLO.show ()
```

```python
from scipy.stats import skew
SKEW = DF_F_E_S ["Working Hours"].skew ()
print(SKEW)
```

```python
Describe_Deadline_Preassure = DF_F_E_S ["Deadline Preassure"].describe ()
print(Describe_Deadline_Preassure)
```

```python
FO = DF_F_E_S [DF_F_E_S ["Working Hours"] > DF_F_E_S ["Working Hours"].quantile (0.93)]
print(FO)
```

```python
from scipy.stats import normaltest
stats , N_T = normaltest (DF_F_E_S ["Working Hours"])
print(N_T)
```

```python
from scipy.stats import skew
import numpy as N
MEAN = N.mean (DF_F_E_S ["Working Hours"])
STD = N.std (DF_F_E_S ["Working Hours"])
OUTLIER = DF_F_E_S [(DF_F_E_S ["Working Hours"] > MEAN + 2 * STD) | (DF_F_E_S ["Working Hours"] < MEAN - 2 * STD)]
print(OUTLIER.skew ())
```
