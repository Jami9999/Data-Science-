```python
D["Fuel Efficiency City"] = 1 / D["Fuel Consumption (City (L/100 km)"]
print(D["Fuel Efficiency City"])
```

```python
print(D.columns)
```

```python
D["Fuel Efficiency Highway"] = 1 / D["Fuel Consumption(Hwy (L/100 km))"]
print(D["Fuel Efficiency Highway"])
```

```python
D["Fuel Efficiency Combined"] = 1 / D["Fuel Consumption(Comb (L/100 km))"]
print(D["Fuel Efficiency Combined"])
```

```python
print(D.columns)
```

```python
S.scatterplot (x = "Engine Size(L)" , y = "Fuel Efficiency City" , data = D)
PLOT.title ("Engine Size vs Fuel Efficiency in City")
PLOT.xlabel ("Engine Size")
PLOT.ylabel ("Fuel Efficiency")
PLOT.show ()
```
