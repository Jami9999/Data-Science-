```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Data_2 = {
    "Club Name" : ["Wendron United FC" , "Wadebridge Town FC" , "AFC Sticker" , "AFC St Day" , "AFC Penzance"] ,
    "League" : ["SWPL Premier Division West" , "SWPL Premier Division West" , "SWPL Premier Division West" , "SWPL Premier Division West" , "SWPL Premier Division West"] ,
    "Tier" : [10 , 10 , 10 , 10 , 10] ,
    "Step" : [6 , 6 , 6 , 6 , 6] ,
    "Country" : ["England" , "England" , "England" , "England" , "England"] ,
    "Manager" : ["Robin Watt" , "Paul Rowe" , "Daniel Nancarrow" , "Joff Mallaber" , None] ,
    "Manager Exists" : [1 , 1 , 1 , 1 , 0] ,
    "Assistant Manager" : [None , "Steve Cudmore" , "Gary Warren" , "Clive Jones" , "Christopher Farley"] ,
    "Assistant Manager Exists" : [0 , 1 , 1 , 1 , 1]
}
DF_2 = P.DataFrame (Data_2)
print(DF_2)
```

```python
X = DF_2 [["Manager Exists" , "Tier" , "Step"]]
Y = DF_2 ["Assistant Manager Exists"]
```

```python
X_Train , X_Test , Y_Train , Y_Test = train_test_split (X , Y , test_size = 0.2 , random_state = 42)
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

```python
M_2 = DecisionTreeClassifier ()
F_2 = M_2.fit (X_Train , Y_Train)
print(F_2)
```

```python
PRE_2 = M_2.predict (X_Test)
print(PRE_2)
```

```python
R2 = r2_score (Y_Test , PRE_2)
print(R2 * 100 , "%")
```
